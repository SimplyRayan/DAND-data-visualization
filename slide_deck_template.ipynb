## Dataset Overview

> the Ford GoBike trip dataset has 170883 rows and 22 columns, such as duration of trip and user ages and type, also weekdays of trips and times of day and other variables that I will be exploring shortly

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df= pd.read_csv('201902-fordgobike-tripdata__Cleaned.csv')

ordered_days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
days_order =pd.api.types.CategoricalDtype(ordered = True,categories=ordered_days)
df['start_weekday'] = df['start_weekday'].astype(days_order)

#  

ordered_times = ['Morning','Afternoon','Evening','Night']
ordered_times =pd.api.types.CategoricalDtype(ordered = True,categories=ordered_times)
df['start_part_of_day'] = df['start_part_of_day'].astype(ordered_times)
```

## Distribution of ages

> Most ages of users are from 25-40 and Distribution is right skewed

```python
plt.figure(figsize=[8, 5])
bins = np.arange(df['age'].min(),df['age'].max()+3,3)
plt.hist(data=df,x='age',bins=bins);
plt.xlabel('Age (years)');
plt.title('Distribution of users ages');
```

## Distribution of trips across weekdays

> most rides are through the work days where the number of rides in the weekends decreases

```python
plt.figure(figsize=[8, 5])
sb.countplot(data=df,x='start_weekday',color='darkblue')
plt.xticks(rotation=15);
plt.xlabel('Weekday')
plt.title('Distribution of trips across weekdays');
```

## Distribution of trips across different time of the data
> Most rides are in the mornings and then in the evening and after noon, with fewer bike rides in the night, this can be the result of users using bikes commute to work



```python
plt.figure(figsize=[8, 5])
sb.countplot(data=df,x='start_part_of_day',color='lightblue')
plt.xticks(rotation=15);
plt.xlabel('Time of day')
plt.title('Distribution of trips across different times of the day');
```

## Duration of trips for different user types 
> the box plot shows that the median duration time of customer's trips is higher than the median of the subscribers trips


```python
plt.figure(figsize=[8, 5])
sb.boxplot(data=df,x='user_type',y='duration_min',color='lightblue');
plt.xlabel('User type')
plt.ylabel('Duration (Mins)')
plt.title('Duration of trips for different user types');
```

## Average duration across weekdays
> the line plot shows that duration of rides on weekends is higher than the duration throug the workdays, I think this can people are busier on work days and don't want to be late, but on weekends they can take their time


```python
day_means= df.groupby('start_weekday').mean().reset_index()

#
plt.figure(figsize=[8, 5])
sb.pointplot(data=day_means,x='start_weekday',y='duration_min',color='purple');
plt.xlabel('Weekday')
plt.ylabel('Duration average (mins)')
plt.title('Average trip duration across the week');
```

## Average users ages across the week
> we can see that average age on workdays is higher than the average age on weekends!


```python
plt.figure(figsize=[8, 5])
sb.pointplot(data=day_means,x='start_weekday',y='age');
plt.xticks(rotation=15);
plt.xlabel('Weekday')
plt.ylabel('Avg age')
plt.title('Average users ages across the week');
```

## Distribution of trips times on different weekdays
> we can see that for all workdays the most common trip times are Morning and Evening but on weekends these two time decreases and trips on the Afternoon increases, again I think this is the result of most people using the bikes to commute to work


```python
stacked_bar = df.groupby(['start_part_of_day', 'start_weekday']).size().reset_index().pivot(columns='start_part_of_day', index='start_weekday', values=0);
stacked_bar.plot(kind='bar', stacked=True,figsize=(8,5));
plt.title('Distribution of trips times on different weekdays');
plt.xlabel('weekday');
```

## Average duration time of different user type across the week
> the plot shows that for subscriber the average druation is almost the same throughout the week, but for the user type customer the average duration on workdays fluctuates between 13-14 mins for, but on weekends it increases to around 16 mins

```python
day_type_means = df.groupby(['start_weekday','user_type']).mean().reset_index()
plt.figure(figsize=[8, 5])
sb.pointplot(data=day_type_means,x='start_weekday',y='duration_min',hue='user_type');
plt.title('Average duration time of different user type across the week')
plt.xlabel('Weekday');
plt.ylabel('Avg duration (min)');
```
